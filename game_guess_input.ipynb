## Number Guessing Game (GUI Version)

```python
import random
import ipywidgets as widgets
from IPython.display import display

# Initialize state
secret_number = random.randint(1, 100)
guesses_taken = 0

# Create UI Elements
title_label = widgets.HTML("<h2>Number Guessing Game</h2>")
instruction_label = widgets.Label("I'm thinking of a number between 1 and 100.")
guess_input = widgets.IntText(value=0, description='Your Guess:')
guess_button = widgets.Button(description="Guess", button_style='primary')
feedback_output = widgets.Output()
play_again_button = widgets.Button(description="Play Again", button_style='success', layout=widgets.Layout(display='none'))

def on_guess_clicked(b):
    global guesses_taken, secret_number
    guesses_taken += 1
    guess = guess_input.value

    with feedback_output:
        feedback_output.clear_output()
        if guess < secret_number:
            print(f"Attempt {guesses_taken}: {guess} is too LOW!")
        elif guess > secret_number:
            print(f"Attempt {guesses_taken}: {guess} is too HIGH!")
        else:
            print(f"Congratulations! {guess} is correct!")
            print(f"You found it in {guesses_taken} guesses.")
            guess_button.disabled = True
            play_again_button.layout.display = 'block'

def on_reset_clicked(b):
    global guesses_taken, secret_number
    secret_number = random.randint(1, 100)
    guesses_taken = 0
    guess_input.value = 0
    guess_button.disabled = False
    play_again_button.layout.display = 'none'
    with feedback_output:
        feedback_output.clear_output()
        print("New game started! Good luck.")

guess_button.on_click(on_guess_clicked)
play_again_button.on_click(on_reset_clicked)

# Layout
ui = widgets.VBox([title_label, instruction_label, guess_input, guess_button, feedback_output, play_again_button])
display(ui)
```
