```python
import random

secret_number = random.randint(1, 100)
attempts = 0

print("ยินดีต้อนรับสู่เกมทายตัวเลข!")
print("ผมได้เลือกตัวเลขระหว่าง 1 ถึง 100 ไว้แล้ว ลองทายดูสิครับ")

# เริ่มต้น Loop การเล่นเกมโดยตรง
while True:
    try:
        guess = int(input("ใส่ตัวเลขที่คุณทาย: "))
        attempts += 1

        if guess < secret_number:
            print(f"{guess} น้อยเกินไป! ลองใหม่อีกครั้ง")
        elif guess > secret_number:
            print(f"{guess} มากเกินไป! ลองใหม่อีกครั้ง")
        else:
            print(f"ยินดีด้วย! คุณทายถูกแล้ว (ใช้ไป {attempts} ครั้ง)")
            print("Congratulations!")
            break
    except ValueError:
        print("กรุณาใส่ตัวเลขที่ถูกต้อง")
```
